We convert VOC's height/width into top-left/bottom-right, and switch x/y coords to be consistent with numpy.

```python
def heightWeightBb(bb): 
    return np.array([bb[1], bb[0], bb[3] + bb[1] - 1, bb[2] + bb[0] - 1])

#defaultdict will simply create any items that you try to access (Python)
trainAnnotations = collections.defaultdict(lambda: [])

for obj in trainJson[ANNOTATIONS]:
    if not obj['ignore']:
        bb = obj[BBOX] #bb coordinates given by pascal
        bb = heightWeightBb(bb) #transformed coordinates
        trainAnnotations[obj[IMG_ID]].append((bb, obj[CAT_ID]))

len(trainAnnotations)
```

Some libs take VOC format bounding boxes, so this let's us convert back when required:

```python
def bbHeightwidth(a): return np.array([a[1] ,a[0] ,a[3] - a[1] + 1, a[2] - a[0] + 1])
```

```python
img = Image.open(IMG_PATH/image0[FILENAME])
```

```python
def showImg(img, figsize=None, ax=None):
    if not ax: fig,ax = plt.subplots(figsize=figsize)
    ax.imshow(img)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    return ax
```

```python
def drawOutline(obj, linewidth):
    obj.set_path_effects([patheffects.Stroke(linewidth=linewidth, foreground='black'), patheffects.Normal()])
```

```python
def drawRectangle(ax, b):
    patch = ax.add_patch(patches.Rectangle(b[:2], *b[-2:], fill=False, edgecolor='white', lw=2))
    drawOutline(patch, 4)
# * is the "splat" operator: 
#      It takes a list as input, and expands it into actual positional arguments in the function call.
```

```python
def drawText(ax, xy, txt, sz=14):
    text = ax.text(*xy, txt, verticalalignment='top', color='white', fontsize=sz, weight='bold')
    drawOutline(text, 1)
```

```python
ax = showImg(img)
b = bbHeightwidth(image0Annotation[0])
drawRectangle(ax, b)
drawText(ax, b[:2], categories[image0Annotation[1]])
```

```python

def drawImage(img, annotation):
    ax = showImg(img, figsize=(16,8))
    for b, c in annotation:
        b = bbHeightwidth(b)
        drawRectangle(ax, b)
        drawText(ax, b[:2], categories[c], sz=14)
```

```python
def drawImageByIndex(i):
    imgAnn = trainAnnotations[i]
    img = Image.open(IMG_PATH/trainFileNames[i])
    #print(img.shape)
    drawImage(img, imgAnn)
```
